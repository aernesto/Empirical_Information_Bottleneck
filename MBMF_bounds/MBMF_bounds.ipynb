```python
hull_names = sort(glob.glob('*.npy'))

hulls = {}
for i,key in enumerate(np.arange(0.0,1.1,.1)):
    if i%2 == 0:
        hulls[key] = np.load(hull_names[i])
        hull = hulls[key]
        plot(hull[:-1,0],hull[:-1,1],'-',label='W: %f'%key)
xlabel('$Ipast$')
ylabel('$Ifuture$')
title('Empirical Bounds - MBMF task')
legend()
```
